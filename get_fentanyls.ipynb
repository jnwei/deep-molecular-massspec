```python
def make_splits_by_family(inchikeys, mainlib_mol_dict, family_names):
    """Separate molecules according to family.
  
    Args:
    inchikeys: List of inchikeys
    mainlib_moldict: dict of lists of rdkit.Mol keyed by inchikey
    family_names: Names of families to filter on.
    Returns:
    dict of list of inchikeys based on families, keyed by family names.
    """
    assert all(f_name in feature_utils.FILTER_DICT.keys() for f_name in family_names)
    inchikey_by_family_dict = dict([(key, []) for key in family_names])
    no_family_list = []

    for inchikey in inchikeys:
        mol = mainlib_mol_dict[inchikey][0]
        for family_key in family_names:
          # TODO(weijennifer): Fix this to account for largest family first. 
          if feature_utils.FILTER_DICT[family_key](mol):
            inchikey_by_family_dict[family_key].append(inchikey)
            break 
        else:
            no_family_list.append(inchikey)

    inchikey_by_family_dict['no_family'] = no_family_list
    return inchikey_by_family_dict
```

```python
class DatasetNameTuple(namedtuple('DatasetNameTuple', ['name', 'family', 'source', 'split'])):
    
    def __new__(cls, name, family, source, split='None'):
        return super(DatasetNameTuple, cls).__new__(cls, name, family, source, split)
    
    def write_base(self):
        if self.name == 'MAINLIB':
            return self.name + '_' + self.family + '_' + self.source
        else:
            return self.name + '_' + self.family + '_' + self.split + '_' + self.source
    
    def write_tfrecord_filename(self):
        return self.write_base() + '.tfrecord'
    
    def write_inchikey_filename(self):
        return self.write_base() + '.inchikey.txt'
```

```python
def get_inchikeys_from_file(filename, base_dir='/mnt/storage/NIST_unzipped/fentanyl_steroid_datasets'):
    with tf.gfile.Open(os.path.join(base_dir, filename)) as f:
        inchikey_list = [line.strip('\n') for line in f.readlines()]
    return inchikey_list

def get_inchikey_list_for_nametuple_list(dataset_name_tuple_list):
    inchikey_list = []
    for dataset_name_tuple in dataset_name_tuple_list:
        sub_inchikey_fname = dataset_name_tuple.write_inchikey_filename()
        sub_inchikey_list = get_inchikeys_from_file(sub_inchikey_fname)
        inchikey_list.extend(sub_inchikey_list)
    return inchikey_list
        
def test_experiment_setup(dataset_setup_dict):
    # test that library matching query not in observed set, or train set
    library_match_query_inchikeys = get_inchikey_list_for_nametuple_list(dataset_setup_dict[LIBRARY_MATCHING_QUERY_KEY])
    library_match_observed_inchikeys = get_inchikey_list_for_nametuple_list(dataset_setup_dict[LIBRARY_MATCHING_OBSERVED_KEY])
    library_match_predicted_inchikeys = get_inchikey_list_for_nametuple_list(dataset_setup_dict[LIBRARY_MATCHING_PREDICTED_KEY])
    spectrum_prediction_inchikeys = get_inchikey_list_for_nametuple_list(dataset_setup_dict[SPECTRUM_PREDICTION_TRAIN_KEY])
    
    assert all(key not in library_match_observed_inchikeys for key in library_match_query_inchikeys)
    assert all(key not in spectrum_prediction_inchikeys for key in library_match_query_inchikeys)
    assert set(library_match_predicted_inchikeys) == set(library_match_query_inchikeys)

    # test that library_matching_predicted and query keys are the same.
```
